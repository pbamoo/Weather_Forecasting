### Missing & Suspect Values

Missing values in time series should not be replaced with the mean, especially if the series is not stationary. 
Some effective alternatives are: Backward Fill, forward-fill the previous value, Linear Interpolation, etc.
The idea is to introduce missing values to the time series, impute it with above approaches and then measure the mean squared error of the imputed against the actual values.
You could also forecast the missing values if you have enough past observations, or backcast the missing values, If you have enough future observations

### Check for & make a Time Series Stationary

A stationary series is one where the values of the series is not a function of time, i.e. the statistical properties of the series like mean, variance and autocorrelation are constant over time. The first step in a forecasting process is typically to do some transformation to convert a non-stationary series to stationary, which makes the forecast more reliable. We can check if a series is stationary by checking the components of the series, i.e trend and seasonality or using the AAugmented Dickey Fuller test(more below) and if not make it stationary via the following:

- Differencing the Series (once or more)
- Take the log of the series
- Take the nth root of the series
- Combination of the above

```python
##1. check the stationarity of the series using the Augmented Dickey Fuller test.
## if p-value < 0.05 significance level ==> stationary 

from statsmodels.tsa.stattools import adfuller
from numpy import log
print("Performing Augmented Dickey-Fuller Test to confirm stationarity...")
result = adfuller(df.temp.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

p = result[1]
if (p > 0.05):
    print("Time Series is NOT Stationary, since p-value > 0.05")
    df = df.diff()  # differencing to make data stationary
else:
    print("Time Series is Stationary, since p-value <= 0.05")
```

#### Components of a time series: 
$$ Base Level + Trend + Seasonality + Error(Random Noise) $$
1. A trend is observed when there is an increasing or decreasing slope observed in the time series. 
2. Seasonality is observed when there is a distinct repeated pattern observed between regular intervals due to seasonal factors.

- A time series can be modeled as an additive or multiplicative, based on the nature of the trend and seasonality.

**Additive time series:**
$$ Value = Base Level + Trend + Seasonality + Error $$

**Multiplicative Time Series:**
$$ Value = Base Level * Trend * Seasonality * Error $$

**ValueError:** Multiplicative seasonality is not appropriate for zero and negative values

(source: https://www.machinelearningplus.com/time-series/time-series-analysis-python/)

```python
#2.
from statsmodels.tsa.seasonal import seasonal_decompose

sdr = seasonal_decompose(df['temp'], model='additive', extrapolate_trend='freq', freq=365)
# multiplicative

sdr.plot();
```

```python
# Extraction of the numerical output of the Components stored in the sdr & put in a dataframe
# Actual Values = Sum of (Seasonal + Trend + Resid)
df_reconstructed = pd.concat([sdr.seasonal, sdr.trend, sdr.resid, sdr.observed], axis=1)
df_reconstructed.columns = ['seasonal', 'trend', 'resid', 'actual_values']
df_reconstructed
```

```python
plt.plot(sdr.seasonal)
plt.plot(sdr.resid,'.')
plt.ylabel('residuals')
plt.xlabel('seasons')
```

```python
# concat the components to dataframe df 
df = pd.concat([df, df_reconstructed], axis=1)
```

#### Decomposing a time series

```python
# Remove the trend by (i) subtracting the trend component obtained from time series decomposition, sdr seen earlier.

detrended = df.temp.values - sdr.trend
plt.plot(detrended)
plt.title('Temperature values detrended by subtracting the trend component', fontsize=12)

# (ii) 
```

```python
# Remove trend and changes in volatility using the percentage difference of the original temperature values.
df['pct_change'] = df['temp'].pct_change()
```

```python
# Remove the seasonality by (i) dividing the series by the seasonal index obtained from the sdr
deseasonalized = df.temp.values / sdr.seasonal
#plt.plot(deseasonalized)
#plt.title('Temperature values deseasonalized', fontsize=12)

# (ii) subtracting the seasonal mean from the value (or the (second order) difference/ percentage_change)
df['seasonal_mean'] = df.groupby('month')['pct_change'].transform('mean')
df['deseasonalized'] = df['pct_change'] - df['seasonal_mean']
```

### ARIMA

An ARIMA model is characterized by 3 terms: p, d, q
* p is the order of the AR term: the number of lags of Y to be used as predictors
* q is the order of the MA term: the number of lagged forecast errors that should go into the ARIMA Model
* d is min the number of differencing required to make the time series stationary, if d=0, time series is already stationary
* p, q values manually by looking at the ACF and PACF plots (below) or via automation via grid search and choose the ARIMA model for which the AIC and BIC values are minimum.

```python
# Autocorrelation and partial autocorrelation functions
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
plot_acf(df.temp.tolist(), lags=30, ax=axes[0])
plot_pacf(df.temp.tolist(), lags=30, ax=axes[1])
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
arima_model = ARIMA(df.temp, order=(2,0,2)).fit(disp=0)
print(arima_model.summary())
```

```python
#1. make prediction using arima_model
predictions = arima_model.predict(start=0, end=len(df)-1)
plt.plot(list(df.temp)[:100], label="Actual")
plt.plot(list(predictions)[:100], 'r', label="Predicted")

plt.xlabel("Time (in Months)")
plt.ylabel("Temperature (C)")
plt.title("Actual and Predicted Temperature Values")

plt.legend(loc='upper center', bbox_to_anchor=(1.45, 0.8))
plt.show()
```

```python
# Plot residual errors
residuals = pd.DataFrame(arima_model.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()

## Residual errors seem fine with near zero mean and uniform variance.
```

```python
# Actual vs Fitted

arima_model.plot_predict(dynamic=False)
plt.show()

# dynamic=False, the in-sample lagged values are used for prediction, i.e. the model gets trained up until the previous value to make the next prediction.
```

```python
# further forecast using Out-of-Time cross-validation, where you take few steps back in time and forecast into the future to as many steps you took back.
```

```python
# Create Training and Test
X = df.temp
train_size = int(len(X) * 0.75) #first 75% for training & remaining 25% for testing
train, test = X[:train_size], X[train_size:]
```

```python
# Steps: Build the ARIMA model on training dataset --> forecast --> plot

# Build Model 
model = ARIMA(train, order=(2, 0, 2)).fit() 

# Forecast
fc, se, conf = model.forecast(6466, alpha=0.05)

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

An alpha of 0.05 means that the ARIMA model will estimate the upper and lower values around the forecast where there is a only a 5% chance that the real value will not be in that range.

```python
## Accuracy Metrics

def forecast_accuracy(forecast, actual):
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return({'me':me, 'mae': mae, 'rmse':rmse,})

forecast_accuracy(fc, test.values)
```

#### Auto Arima

```python
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm

model = pm.auto_arima(df.temp, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)

print(model.summary())
```

The Akaike Information Critera (AIC) is a widely used measure of a statistical model. It basically quantifies 1) the goodness of fit, and 2) the simplicity/parsimony, of the model into a single statistic. (https://coolstatsblog.com/2013/08/14/using-aic-to-test-arima-models-2)

```python
# Forecast
n_periods = 365
fc, confint = model.predict(n_periods=n_periods, return_conf_int=True)
index_of_fc = np.arange(len(df.temp), len(df.temp)+n_periods)

# make series for plotting purpose
fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)

# Plot
plt.plot(df.temp)
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("Final Forecast Temperature")
plt.show()
```
